```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
premise_file = "listings.csv"
airbnb_df = pd.read_csv(premise_file)
airbnb_df.head()
```

```python
relevant_columns = ["name", "neighbourhood_group", "neighbourhood", "latitude", 
                    "longitude", "room_type", "price", "minimum_nights", "availability_365"]
airbnb_relevant = airbnb_df[relevant_columns].copy()

airbnb_relevant.index.name = "ID"
airbnb_relevant.head()
```

```python
connection_string = "postgres:bangladesh@localhost:5432/airbnb_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
airbnb_relevant.to_sql(name='airbnb', con=engine, if_exists='append', index=True)
```
